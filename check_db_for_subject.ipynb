```python
from bcpp_subject.models import SubjectVisit, HivCareAdherence

missing_results = {}
file_name = '/home/django/source/check18.csv'
f = open(file_name, 'r')
lines = f.readlines()
lines.pop(0)

for line in lines:
    line = line.strip()
    subject_identifier, visit_code = line.split(',')
    missing_results[subject_identifier] = visit_code
```

```python
machine_db = []
existing_hiv_care = []
for subject_identifier, visit_code in missing_results.items():
    try:
        hiv_ca = HivCareAdherence.objects.get(subject_visit__subject_identifier=subject_identifier, subject_visit__visit_code=visit_code)
    except HivCareAdherence.DoesNotExist:
        try:
            sv = SubjectVisit.objects.get(subject_identifier=subject_identifier, visit_code=visit_code)
        except SubjectVisit.DoesNotExist:
            pass
        else:
            map_area = sv.household_member.household_structure.household.plot.map_area
            machine_db.append([subject_identifier, visit_code, sv.created, sv.hostname_created, map_area])
    else:
        existing_hiv_care.append([subject_identifier, visit_code])
print(len(existing_hiv_care), 'Total that are on the server')
print(len(machine_db), 'Total visits found')
print(len(missing_results), 'Total missing')
import pandas as pd
df = pd.DataFrame(machine_db)
df.df.to_csv('hiv_care_adherence.csv', encoding='utf-8')
```
