```python
from bcpp_subject.models import SubjectVisit

missing_results = {}
file_name = '/Users/ckgathi/source/check5.csv'
f = open(file_name, 'r')
lines = f.readlines()
lines.pop(0)

for line in lines:
    line = line.strip()
    map_area, subject_identifier, visit_code = line.split(',')
    missing_results[subject_identifier] = [map_area, visit_code]
```

```python
machine_db = []
for subject_identifier, dt in missing_results.items():
    visit_code=dt[1]
    map_area = dt[0]
    try:
        sv = SubjectVisit.objects.using('bhp066').get(subject_identifier=subject_identifier, visit_code=visit_code)
    except SubjectVisit.DoesNotExist:
        pass
    else:
        try:
            d = Demographics.objects.get(subject_visit=sv)
        except Demographics.DoesNotExist:
            machine_db.append([subject_identifier, visit_code, sv.created, sv.hostname_created, map_area])
        else:
            machine_db.append([subject_identifier, visit_code, d.created, d.hostname_created, map_area])
```

```python
for mdb in machine_db:
    print(mdb[0], mdb[1], mdb[2].strftime("%Y-%m-%d %H:%M:%S"), mdb[3], mdb[4])
```
